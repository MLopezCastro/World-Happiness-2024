# 🌍 World Happiness 2024 Report Analysis
### Análisis del Informe Mundial de la Felicidad 2024

Este proyecto analiza los datos del **World Happiness Report**, que mide la felicidad de los países a través de diferentes indicadores como el PIB per cápita, apoyo social, expectativa de vida saludable, libertad para tomar decisiones, generosidad y percepciones de corrupción.

El objetivo de este proyecto es:
- Explorar qué factores influyen más en la felicidad de los países.
- Visualizar las tendencias de la felicidad a lo largo de los años.
- Crear un portafolio que muestre habilidades en **Python**, **SQL** y **Tableau**.

---

## 📊 Dataset: World Happiness Report
**Fuente:** [World Happiness Report](https://worldhappiness.report)  

### 📂 Variables Incluidas:
- `Country name`: Nombre del país.
- `Ladder score`: Puntuación de felicidad en una escala del 0 al 10.
- `upperwhisker`: Límite superior del intervalo de confianza.
- `lowerwhisker`: Límite inferior del intervalo de confianza.
- `Explained by: Log GDP per capita`: Parte de la felicidad explicada por el PIB per cápita.
- `Explained by: Social support`: Parte de la felicidad explicada por el apoyo social.
- `Explained by: Healthy life expectancy`: Parte de la felicidad explicada por la esperanza de vida saludable.
- `Explained by: Freedom to make life choices`: Parte de la felicidad explicada por la libertad para tomar decisiones.
- `Explained by: Generosity`: Parte de la felicidad explicada por la generosidad.
- `Explained by: Perceptions of corruption`: Parte de la felicidad explicada por la percepción de corrupción.
- `Dystopia + residual`: Puntuación de referencia que representa cuánto peor sería un país hipotético distópico en comparación con el país real.

# Importación de Librerías y Carga de Datos #

```python
# Importar las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo de gráficos
sns.set(style="whitegrid")

# Cargar el primer archivo: DataForTable2.1.csv (separado por punto y coma)
data_table = pd.read_csv(
    r'C:\Users\mlope\OneDrive\Escritorio\MLopezCastro\World Happiness 2024\Files\DataForTable2.1.csv',
    sep=';',
    encoding='windows-1252'
)

# Cargar el segundo archivo: DataForFigure2.1.csv (separado por punto y coma)
data_figure = pd.read_csv(
    r'C:\Users\mlope\OneDrive\Escritorio\MLopezCastro\World Happiness 2024\Files\DataForFigure2.1.csv',
    sep=';',
    encoding='windows-1252'
)

# Mostrar las primeras filas de ambos datasets
print("Primer archivo (DataForTable2.1):")
display(data_table.head())

print("\nSegundo archivo (DataForFigure2.1):")
display(data_figure.head())
```

## Exploración Inicial de los Datos ##

```python
# 📋 Información general del primer archivo (DataForTable2.1)
print("🔍 Información del archivo DataForTable2.1:")
data_table.info()

# Verificar valores faltantes en DataForTable2.1
print("\n❓ Valores faltantes en DataForTable2.1:")
print(data_table.isnull().sum())

# 📋 Información general del segundo archivo (DataForFigure2.1)
print("\n🔍 Información del archivo DataForFigure2.1:")
data_figure.info()

# Verificar valores faltantes en DataForFigure2.1
print("\n❓ Valores faltantes en DataForFigure2.1:")
print(data_figure.isnull().sum())

# 📊 Estadísticas descriptivas para ambos datasets
print("\n📊 Estadísticas descriptivas del archivo DataForTable2.1:")
display(data_table.describe())

print("\n📊 Estadísticas descriptivas del archivo DataForFigure2.1:")
display(data_figure.describe())
```

## 📌 Manejo de Valores Faltantes (NaN)

Durante la exploración inicial de los datos, hemos identificado algunos valores faltantes en las columnas de ambos archivos cargados (`DataForTable2.1` y `DataForFigure2.1`).

Por ahora, hemos decidido **ignorar los valores faltantes** en lugar de imputarlos. Esto significa que cualquier análisis o visualización que realicemos se basará únicamente en los datos disponibles, omitiendo las filas que contengan valores `NaN`.

En futuras etapas del proyecto, podríamos considerar métodos más avanzados para manejar los valores faltantes, como:
- Rellenar los valores faltantes con la media o mediana.
- Utilizar métodos de interpolación.

Para este análisis inicial, utilizaremos la función **`dropna()`** de Pandas para asegurarnos de que los valores faltantes no afecten nuestros gráficos ni resultados.


------------

## 🧑‍💻 Ajuste en el Código: Filtrar los Valores Faltantes (NaN) ##

```python
# Eliminar valores faltantes antes de hacer cualquier análisis
clean_data_table = data_table.dropna()
clean_data_figure = data_figure.dropna()

# Mostrar las primeras filas de los datasets limpios
print("📋 Primer archivo limpio (DataForTable2.1):")
display(clean_data_table.head())

print("\n📋 Segundo archivo limpio (DataForFigure2.1):")
display(clean_data_figure.head())
```

------------------

## Análisis Visual Exploratorio (EDA) ##

```python
# 🥇 Gráfico 1: Top 10 Países Más Felices (Datos de 2023, Reporte 2024)
top_10_happiest = clean_data_table[clean_data_table['year'] == 2023].sort_values(by='Life Ladder', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Life Ladder', y='Country name', data=top_10_happiest, palette='viridis')
plt.title('🌍 Top 10 Países Más Felices del Mundo (Datos de 2023, Reporte 2024)')
plt.xlabel('Puntuación de Felicidad (Life Ladder)')
plt.ylabel('País')
plt.show()
```

## Tendencia de la Felicidad en Finlandia ##

```python
# 📈 Gráfico 2: Tendencia de la Felicidad en Finlandia (2008-2023, Reporte 2024)
finland_trend = clean_data_table[clean_data_table['Country name'] == 'Finland']

plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='Life Ladder', data=finland_trend, marker='o')
plt.title('📈 Tendencia de la Felicidad en Finlandia (2008-2023, Reporte 2024)')
plt.xlabel('Año')
plt.ylabel('Puntuación de Felicidad')
plt.xticks(finland_trend['year'])
plt.show()
```
