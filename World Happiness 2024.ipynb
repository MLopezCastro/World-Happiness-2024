# üåç World Happiness 2024 Report Analysis
### An√°lisis del Informe Mundial de la Felicidad 2024

Este proyecto analiza los datos del **World Happiness Report**, que mide la felicidad de los pa√≠ses a trav√©s de diferentes indicadores como el PIB per c√°pita, apoyo social, expectativa de vida saludable, libertad para tomar decisiones, generosidad y percepciones de corrupci√≥n.

## üìä Dataset: World Happiness Report

### üìÇ Variables Incluidas:
- `Country name`: Nombre del pa√≠s.
- `Ladder score`: Puntuaci√≥n de felicidad en una escala del 0 al 10.
- `upperwhisker`: L√≠mite superior del intervalo de confianza.

- `Explained by: Log GDP per capita`: Parte de la felicidad explicada por el PIB per c√°pita.

- `Dystopia + residual`: Puntuaci√≥n de referencia que representa cu√°nto peor ser√≠a un pa√≠s hipot√©tico dist√≥pico en comparaci√≥n con el pa√≠s real.

# Importaci√≥n de Librer√≠as y Carga de Datos #

```python
# Importar las librer√≠as necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo de gr√°ficos
sns.set(style="whitegrid")

# Cargar el primer archivo: DataForTable2.1.csv (separado por punto y coma)
data_table = pd.read_csv(
    r'C:\Users\mlope\OneDrive\Escritorio\MLopezCastro\World Happiness 2024\Files\DataForTable2.1.csv',
    sep=';',
    encoding='windows-1252'
)

# Cargar el segundo archivo: DataForFigure2.1.csv (separado por punto y coma)
data_figure = pd.read_csv(
    r'C:\Users\mlope\OneDrive\Escritorio\MLopezCastro\World Happiness 2024\Files\DataForFigure2.1.csv',
    sep=';',
    encoding='windows-1252'
)

# Mostrar las primeras filas de ambos datasets
print("Primer archivo (DataForTable2.1):")
display(data_table.head())

print("\nSegundo archivo (DataForFigure2.1):")
display(data_figure.head())
```

-----------------

## Exploraci√≥n Inicial de los Datos ##

```python
# üìã Informaci√≥n general del primer archivo (DataForTable2.1)
print("üîç Informaci√≥n del archivo DataForTable2.1:")
data_table.info()

# Verificar valores faltantes en DataForTable2.1
print("\n‚ùì Valores faltantes en DataForTable2.1:")
print(data_table.isnull().sum())

# üìã Informaci√≥n general del segundo archivo (DataForFigure2.1)
print("\nüîç Informaci√≥n del archivo DataForFigure2.1:")
data_figure.info()

# Verificar valores faltantes en DataForFigure2.1
print("\n‚ùì Valores faltantes en DataForFigure2.1:")
print(data_figure.isnull().sum())

# üìä Estad√≠sticas descriptivas para ambos datasets
print("\nüìä Estad√≠sticas descriptivas del archivo DataForTable2.1:")
display(data_table.describe())

print("\nüìä Estad√≠sticas descriptivas del archivo DataForFigure2.1:")
display(data_figure.describe())
```

--------------------

## üìå Manejo de Valores Faltantes (NaN)

Durante la exploraci√≥n inicial de los datos, hemos identificado algunos valores faltantes en las columnas de ambos archivos cargados (`DataForTable2.1` y `DataForFigure2.1`).

Por ahora, hemos decidido **ignorar los valores faltantes** en lugar de imputarlos. Esto significa que cualquier an√°lisis o visualizaci√≥n que realicemos se basar√° √∫nicamente en los datos disponibles, omitiendo las filas que contengan valores `NaN`.

En futuras etapas del proyecto, podr√≠amos considerar m√©todos m√°s avanzados para manejar los valores faltantes, como:
- Rellenar los valores faltantes con la media o mediana.
- Utilizar m√©todos de interpolaci√≥n.

Para este an√°lisis inicial, utilizaremos la funci√≥n **`dropna()`** de Pandas para asegurarnos de que los valores faltantes no afecten nuestros gr√°ficos ni resultados.


------------

## üßë‚Äçüíª Ajuste en el C√≥digo: Filtrar los Valores Faltantes (NaN) ##

```python
# Eliminar valores faltantes antes de hacer cualquier an√°lisis
clean_data_table = data_table.dropna()
clean_data_figure = data_figure.dropna()

# Mostrar las primeras filas de los datasets limpios
print("üìã Primer archivo limpio (DataForTable2.1):")
display(clean_data_table.head())

print("\nüìã Segundo archivo limpio (DataForFigure2.1):")
display(clean_data_figure.head())
```

------------------

## An√°lisis Visual Exploratorio (EDA) ##

```python
# ü•á Gr√°fico 1: Top 10 Pa√≠ses M√°s Felices (Datos de 2023, Reporte 2024)
top_10_happiest = clean_data_table[clean_data_table['year'] == 2023].sort_values(by='Life Ladder', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Life Ladder', y='Country name', data=top_10_happiest, palette='viridis')
plt.title('üåç Top 10 Pa√≠ses M√°s Felices del Mundo (Datos de 2023, Reporte 2024)')
plt.xlabel('Puntuaci√≥n de Felicidad (Life Ladder)')
plt.ylabel('Pa√≠s')
plt.show()
```

## Tendencia de la Felicidad en un pa√≠s a lo largo de los a√±os (por ejemplo, Finlandia) ##

```python
# üìà Gr√°fico 2: Tendencia de la Felicidad en Finlandia (2008-2023, Reporte 2024)
finland_trend = clean_data_table[clean_data_table['Country name'] == 'Finland']

plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='Life Ladder', data=finland_trend, marker='o')
plt.title('üìà Tendencia de la Felicidad en Finlandia (2008-2023, Reporte 2024)')
plt.xlabel('A√±o')
plt.ylabel('Puntuaci√≥n de Felicidad')
plt.xticks(finland_trend['year'])
plt.show()
```

```python
# Lista de pa√≠ses para analizar
paises = ['Argentina', 'Brazil', 'United States', 'Canada', 'Spain', 'Australia']

# Crear un gr√°fico de l√≠nea para cada pa√≠s
for pais in paises:
    # Filtrar los datos para el pa√≠s actual
    pais_trend = clean_data_table[clean_data_table['Country name'] == pais]
    
    # Crear el gr√°fico
    plt.figure(figsize=(12, 6))
    sns.lineplot(x='year', y='Life Ladder', data=pais_trend, marker='o')
    plt.title(f'üìà Tendencia de la Felicidad en {pais} (2008-2023, Reporte 2024)')
    plt.xlabel('A√±o')
    plt.ylabel('Puntuaci√≥n de Felicidad')
    plt.xticks(pais_trend['year'])
    plt.show()
```

```python
# Verificar los tipos de datos de las columnas
print(clean_data_table.dtypes)

# Ver los primeros valores de las columnas relevantes
print(clean_data_table[['Life Ladder', 'Log GDP per capita', 'Social support']].head())
```

Las columnas que deber√≠an ser num√©ricas est√°n en formato object (texto) y usan comas (,) en lugar de puntos (.) como separador decimal.

-------------

‚úÖ Paso 1: Convertir las Comas en Puntos y Cambiar a Tipo Num√©rico

```python
# Convertir columnas num√©ricas que est√°n en formato texto
columnas_a_convertir = [
    'Life Ladder',
    'Log GDP per capita',
    'Social support',
    'Healthy life expectancy at birth',
    'Freedom to make life choices',
    'Generosity',
    'Perceptions of corruption',
    'Positive affect',
    'Negative affect'
]

# Reemplazar comas por puntos y convertir a float
for columna in columnas_a_convertir:
    clean_data_table[columna] = clean_data_table[columna].str.replace(',', '.').astype(float)
```

------------

‚úÖ Paso 2: Verificar que la Conversi√≥n Funcione

```python
# Verificar los nuevos tipos de datos
print(clean_data_table.dtypes)
```

## Matriz de Correlaci√≥n (Heatmap) ##

```python
# üîç Gr√°fico 3: Matriz de Correlaci√≥n de Factores de Felicidad
plt.figure(figsize=(12, 8))

# Seleccionar las columnas relevantes para la correlaci√≥n
columnas_relevantes = [
    'Life Ladder',
    'Log GDP per capita',
    'Social support',
    'Healthy life expectancy at birth',
    'Freedom to make life choices',
    'Generosity',
    'Perceptions of corruption',
    'Positive affect',
    'Negative affect'
]

# Filtrar los datos para esas columnas
numerical_data = clean_data_table[columnas_relevantes]

# Calcular la matriz de correlaci√≥n
correlation = numerical_data.corr()

# Crear el heatmap
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('üîó Matriz de Correlaci√≥n de Factores de Felicidad')
plt.show()
```
