# 🌍 World Happiness 2024 Report Analysis
### Análisis del Informe Mundial de la Felicidad 2024

Este proyecto analiza los datos del **World Happiness Report**, que mide la felicidad de los países a través de diferentes indicadores como el PIB per cápita, apoyo social, expectativa de vida saludable, libertad para tomar decisiones, generosidad y percepciones de corrupción.

El objetivo de este proyecto es:
- Explorar qué factores influyen más en la felicidad de los países.
- Visualizar las tendencias de la felicidad a lo largo de los años.
- Crear un portafolio que muestre habilidades en **Python**, **SQL** y **Tableau**.

---

## 📊 Dataset: World Happiness Report
**Fuente:** [World Happiness Report](https://worldhappiness.report)  

### 📂 Variables Incluidas:
- `Country name`: Nombre del país.
- `Ladder score`: Puntuación de felicidad en una escala del 0 al 10.
- `upperwhisker`: Límite superior del intervalo de confianza.
- `lowerwhisker`: Límite inferior del intervalo de confianza.
- `Explained by: Log GDP per capita`: Parte de la felicidad explicada por el PIB per cápita.
- `Explained by: Social support`: Parte de la felicidad explicada por el apoyo social.
- `Explained by: Healthy life expectancy`: Parte de la felicidad explicada por la esperanza de vida saludable.
- `Explained by: Freedom to make life choices`: Parte de la felicidad explicada por la libertad para tomar decisiones.
- `Explained by: Generosity`: Parte de la felicidad explicada por la generosidad.
- `Explained by: Perceptions of corruption`: Parte de la felicidad explicada por la percepción de corrupción.
- `Dystopia + residual`: Puntuación de referencia que representa cuánto peor sería un país hipotético distópico en comparación con el país real.

## Calcular la correlación entre dos variables (por ejemplo, PIB per cápita y felicidad) ##
Primero, vamos a calcular el coeficiente de correlación de Pearson entre Log GDP per capita y Life Ladder (felicidad general).

```python
# Calcular la correlación de Pearson entre Log GDP per capita y Life Ladder
correlacion_pib_felicidad = clean_data_table['Log GDP per capita'].corr(clean_data_table['Life Ladder'])

# Mostrar el resultado
print(f"📊 Correlación entre PIB per cápita y felicidad: {correlacion_pib_felicidad:.2f}")
```

```python
# Calcular correlaciones para diferentes variables
correlacion_social = clean_data_table['Social support'].corr(clean_data_table['Life Ladder'])
correlacion_vida = clean_data_table['Healthy life expectancy at birth'].corr(clean_data_table['Life Ladder'])
correlacion_libertad = clean_data_table['Freedom to make life choices'].corr(clean_data_table['Life Ladder'])

# Mostrar los resultados
print(f"🤝 Correlación entre apoyo social y felicidad: {correlacion_social:.2f}")
print(f"🧬 Correlación entre esperanza de vida y felicidad: {correlacion_vida:.2f}")
print(f"🗽 Correlación entre libertad de elección y felicidad: {correlacion_libertad:.2f}")
```

1. Reemplazamos las **comas por puntos** en las columnas numéricas.
2. Convertimos las columnas al tipo **float** (números decimales).

## 📊 Variación de Felicidad por País (2008-2023) ##

En este análisis, calculamos cómo ha cambiado la puntuación de felicidad (`Life Ladder`) en cada país entre los años 2008 y 2023. Mostramos un ranking de los países que más mejoraron y los que más empeoraron durante este período.


```python
# Calcular la felicidad promedio por país en 2008 y 2023
felicidad_pais_2008 = clean_data_table[clean_data_table['year'] == 2008].set_index('Country name')['Life Ladder']
felicidad_pais_2023 = clean_data_table[clean_data_table['year'] == 2023].set_index('Country name')['Life Ladder']

# Calcular la variación de felicidad por país
variacion_felicidad_pais = (felicidad_pais_2023 - felicidad_pais_2008).dropna().sort_values(ascending=False)

# Mostrar los 10 países que más mejoraron y empeoraron
print("📈 Top 10 Países que Más Mejoraron:")
print(variacion_felicidad_pais.head(10))
print("\n📉 Top 10 Países que Más Empeoraron:")
print(variacion_felicidad_pais.tail(10))
```
